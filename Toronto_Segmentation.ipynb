```python
# Libraries
import requests
import lxml.html as lh
import bs4 as bs
import urllib.request
import numpy as np 
import pandas as pd

import folium
from geopy.geocoders import Nominatim
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
url
```

```python
df = requests.get(url)
df
```

```python
#Creating Beutiful Soup Obeject from df
sop = bs.BeautifulSoup(df.content,'lxml')
sop
```

```python
df1 = sop.find_all('table')[0]
df1 = str(df1)
df1
```

```python
df2 = pd.read_html(df1)
df2
```

```python
dff = pd.read_json(df2[0].to_json(orient='records'))
dff
```

```python
# drop cells with not assigned borough
dff = dff[dff.Borough != "Not assigned"].reset_index(drop=True)
dff
```

```python
#Groupby
dff = dff.groupby(['Borough', 'Postal Code'], as_index=False).agg(','.join)
dff
```

```python
#Replacing not assignmed values in Neighbourhood with Borough
dff['Neighbourhood'] = np.where(dff['Neighbourhood'] == 'Not assigned', dff['Borough'], dff['Neighbourhood'])
```

```python
dff
```

```python
data = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
data
```

```python
# merge
dff_1 = dff.merge(data, on="Postal Code", how="left")
dff_1.head()
```

```python
# Changing Column name
dff_1 = dff_1.rename(columns={"Postal Code": "PostalCode"})
dff_1
```

# Task 3

#### Making dataframe using Boroughs that contain word Toronto

```python
df_t = dff_1[dff_1['Borough'].str.contains('Toronto')]
```

```python
df_t
```

#### Map of Toronto

```python
map_toronto = folium.Map(location=[43.728020, -79.388790], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_t['Latitude'], df_t['Longitude'], df_t['Borough'], df_t['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_toronto)  
    
map_toronto
```
